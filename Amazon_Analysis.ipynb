## 2. Data Cleaning

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '')
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '')
```

```python
df['discounted_price'] = df['discounted_price'].astype(float)
df['actual_price'] = df['actual_price'].astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = df['rating_count'].astype(float)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['rating', 'rating_count'])
```

```python
df.isnull().sum()
```

```python
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

## 4. Insights

```python
# INSIGHTS

# 1. Categories with the highest ratings:
# Tablets in Computer&Accessories category tend to have higher average ratings, likely due to more targeted customer expectations and product use cases.

# 2. Discount vs Rating:
# There is a weak negative relationship between discount percentage and product ratings, suggesting that higher discounts do not translate to better customer satisfaction. 
# This implies that product quality, rather than pricing strategy, is the primary driver of ratings.

# 3. Popularity drivers:
# Product popularity (measured by review count) is not strongly influenced by price, discount levels, or even ratings. 
# This suggests that external factors such as brand recognition, product visibility, and market demand play a more significant role in driving engagement.

# 4. High Engagement (Review Count)
# Mass-market consumer electronics dominate customer engagement, indicating high demand and frequent purchase behavior in these categories.
```

## 5. Recommendation System

```python
recommended = df[
    (df['rating'] >= 4.0) &
    (df['rating_count'] >= 1000) &
    (df['discount_percentage'] >= 20)
]
```

```python
recommended = recommended.sort_values(by='rating', ascending=False)

recommended[['product_name', 'rating', 'rating_count', 'discount_percentage']].head(5)
```

```python
# CONCLUSION

# - High ratings do not strongly depend on discounts
# - Popularity is influenced by external factors such as brand recognition, product visibility, and market demand plays a more significant role in driving engagement.
# - Top categories are: Tablets, Adapter, Film, StreamingClients and Basic Category
# - Recommended products were identified using the sort by rating method
```

## Final Insights

- High ratings are not strongly influenced by discounts  
- Popularity is driven more by demand than pricing or ratings  
- Certain categories show consistently higher customer satisfaction  
- A combination of rating, engagement, and discount can identify high-value products
